# GEOMETRIC MEAN

```python
y = 1, 3, 9, 27, 81, 243, 729
```

```python
mean = np.mean(y)
mean
```

```python
import pandas as pd
from matplotlib import pyplot as plt
range1=0,1,2,3,4,5,6
y=1,3,25,56,89,256,953
mean=[156.14 for i in range(len(range1))]
plt.plot(range1, y,marker='*')
plt.plot(mean)

plt.grid()
plt.xlabel('range)')
plt.ylabel('YYYYYYYYYYY')
```

```python
from scipy.stats.mstats import gmean 
gm = gmean(y)
gm
```
